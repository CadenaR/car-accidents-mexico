```python
import pandas as pd
import numpy as np
acc1 = pd.read_csv("./atus_anual_2009.csv", index_col=False)
acc2 = pd.read_csv("./atus_anual_2010.csv", index_col=False)
acc3 = pd.read_csv("./atus_anual_2011.csv", index_col=False)
acc4 = pd.read_csv("./atus_anual_2012.csv", index_col=False)
acc5 = pd.read_csv("./atus_anual_2013.csv", index_col=False)
acc6 = pd.read_csv("./atus_anual_2014.csv", index_col=False)
acc7 = pd.read_csv("./atus_anual_2015.csv", index_col=False)
acc8 = pd.read_csv("./atus_anual_2016.csv", index_col=False)
acc9 = pd.read_csv("./atus_anual_2017.csv", index_col=False)
acc10 = pd.read_csv("./atus_anual_2018.csv", index_col=False)
acc11 = pd.read_csv("./atus_anual_2019.csv", index_col=False)
```

```python
df = acc1.append(acc2, ignore_index = True)
df = df.append(acc3, ignore_index = True)
df = df.append(acc4, ignore_index = True)
df = df.append(acc5, ignore_index = True)
df = df.append(acc6, ignore_index = True)
df = df.append(acc7, ignore_index = True)
df = df.append(acc8, ignore_index = True)
df = df.append(acc9, ignore_index = True)
df = df.append(acc10, ignore_index = True)
df = df.append(acc11, ignore_index = True)
```

```python
df["REGION"] = df["ID_ENTIDAD"].astype(str) + "-" + df["ID_MUNICIPIO"].astype(str)
df = df.drop(columns=['COBERTURA', 'ID_MINUTO', 'ID_DIA', 'ESTATUS', 'ID_ENTIDAD', 'ID_MUNICIPIO', 'MICROBUS', 'OMNIBUS', 'TRANVIA', 'TRACTOR', 'FERROCARRI', 'CONDMUERTO', 'PASAMUERTO', 'PEATMUERTO', 'CICLMUERTO', 'OTROMUERTO', 'PEATHERIDO', 'CICLHERIDO', 'OTROHERIDO', 'NEMUERTO', 'NEHERIDO', 'URBANA', 'SUBURBANA', 'CAMPASAJ', 'PASCAMION', 'CAMIONETA', 'CAMION', 'MOTOCICLET', 'BICICLETA', 'OTROVEHIC', 'CAPAROD', 'PASAHERIDO'])
df["DIASEMANA"] = df['DIASEMANA'].str.lower().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df["TIPACCID"] = df['TIPACCID'].str.lower().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df["CAUSAACCI"] = df['CAUSAACCI'].str.lower().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df["SEXO"] = df['SEXO'].str.lower().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df["ALIENTO"] = df['ALIENTO'].str.lower().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df["CINTURON"] = df['CINTURON'].str.lower().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df["CLASACC"] = df['CLASACC'].str.lower().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df["TIPACCID"].replace({"colision con vehculo automotor": "colision con vehiculo automotor", "cada de pasajero": "caida de pasajero", "colisin con animal": "colision con animal", "colisin con ciclista": "colision con ciclista", "colisin con ferrocarril": "colision con ferrocarril", "colisin con motocicleta": "colision con motocicleta", "colisin con objeto fijo": "colision con objeto fijo", "colisin con peatn (atropellamiento)": "colision con peaton (atropellamiento)", "colisin con vehculo automotor": "colision con vehculo automotor"}, inplace=True)
df["SEXO"].replace({"se fug": "se fugo"}, inplace=True)
df["ALIENTO"].replace({"s": "si"}, inplace=True)
df["CINTURON"].replace({"s": "si"}, inplace=True)
df["CAUSAACCI"].replace({"falla del vehculo": "falla del vehiculo", "mala condicin del camino": "mala condicion del camino", "peatn o pasajero": "peaton o pasajero"}, inplace=True)
df["CLASACC"].replace({"slo daos": "solo danos"}, inplace=True)
df = df[df.SEXO != "Certificado cero"]
df = df.drop(df[(df.DIASEMANA == 'certificado cero')].index)
df.head()
```
