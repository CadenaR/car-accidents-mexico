```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import plot_confusion_matrix

%matplotlib inline

df = pd.read_csv("./atus_2009_2019.csv", index_col=False)
```

```python
for col in df:
  print(col)
  print(len(df[col].unique()))
  print(df[col].unique())
  print()
```

```python
df = df.drop(columns=['MES', 'ANIO'])

df = df[df.DIASEMANA != "certificado cero"]
df = df[df.TIPACCID != "certificado cero"]
df = df[df.ID_HORA != 99]

df.REGION = pd.factorize(df.REGION)[0]

df.DIASEMANA = pd.factorize(df.DIASEMANA)[0]

df.TIPACCID = pd.factorize(df.TIPACCID)[0]

df.CAUSAACCI = pd.factorize(df.CAUSAACCI)[0]

df.SEXO = pd.factorize(df.SEXO)[0]

df.ALIENTO = pd.factorize(df.ALIENTO)[0]

df.CINTURON = pd.factorize(df.CINTURON)[0]

df.CLASACC = pd.factorize(df.CLASACC)[0]
```

```python
for col in df:
  print(col)
  print(len(df[col].unique()))
  print(df[col].unique())
  print()
```

```python
df.to_csv('atus_2009_2019_clean.csv',index=False)
```
